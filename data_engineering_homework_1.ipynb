```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('green_tripdata_2019-10.csv')
```

```python
df_2 = pd.read_csv('taxi_zone_lookup.csv')
```

```python
df.head()
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'])
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'])
```

```python
df_2.head()
```

```python
df.shape
```

```python
filtered_data = df[(df['lpep_pickup_datetime'] >= '2019-10-01') & (df['lpep_pickup_datetime'] < '2019-11-01')]

conditions = [
    (filtered_data['trip_distance'] <= 1),
    (filtered_data['trip_distance'] > 1) & (filtered_data['trip_distance'] <= 3),
    (filtered_data['trip_distance'] > 3) & (filtered_data['trip_distance'] <= 7),
    (filtered_data['trip_distance'] > 7) & (filtered_data['trip_distance'] <= 10),
    (filtered_data['trip_distance'] > 10)
]

segment_counts = [filtered_data[condition].shape[0] for condition in conditions]
```

```python
segment_counts
```

```python
filtered_data['pickup_date'] = filtered_data['lpep_pickup_datetime'].dt.date
longest_trip_per_day = filtered_data.loc[filtered_data.groupby('pickup_date')['trip_distance'].idxmax()]
longest_trip_day = longest_trip_per_day.loc[longest_trip_per_day['trip_distance'].idxmax()]

print(f"Longest Trip Day. {longest_trip_day['pickup_date']}.")
```

```python
oct_18_data = df[df['lpep_pickup_datetime'].dt.date == pd.to_datetime('2019-10-18').date()]
top_pickup_zones = oct_18_data.groupby('PULocationID').agg({'total_amount': 'sum'})
top_pickup_zones = top_pickup_zones[top_pickup_zones['total_amount'] > 13000].sort_values(by='total_amount', ascending=False)

print(f"Top Pickup Zones on 2019-10-18. {top_pickup_zones}")
```

```python
merged_df = pd.merge(df, df_2, left_on='PULocationID', right_on='LocationID')
filtered_df = merged_df[merged_df['lpep_pickup_datetime'].dt.date == pd.to_datetime('2019-10-18').date()]
locations_sum = filtered_df.groupby('Zone')['total_amount'].sum()
top_locations = locations_sum.sort_values(ascending=False).head(3)
print(top_locations)
```

```python
zone_data = filtered_df[filtered_df['Zone'] == "East Harlem North"]
largest_tip_zone = zone_data.loc[zone_data['tip_amount'].idxmax()]
zone_with_largest_tip = largest_tip_zone['DOLocationID']

print(f"Zone with Largest Tip for East Harlem North. {zone_with_largest_tip}.")
```
